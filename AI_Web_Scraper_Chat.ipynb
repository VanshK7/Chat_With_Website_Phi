```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Splitting the text into chunks of 1000 with an overlap of 50
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
documents = text_splitter.split_documents(text_documents)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

# Load the HuggingFace embeddings
hf_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Parameters to control LLM's response generation
generation_args = {
    "max_new_tokens": 512,          # Maximum length of the response
    "return_full_text": False,      # Only return the generated text
}
```

```python
# Create a FAISS vector store from the text chunks
from langchain.vectorstores import FAISS

vector_store = FAISS.from_texts(texts=[doc.page_content for doc in documents], embedding=hf_embeddings)
```

```python
# Importing the necessary libraries
from langchain.prompts import ChatPromptTemplate
from langchain.chains import ConversationalRetrievalChain
```

```python
# Defining the function to interact with the context
def search_context(query):
    results = vector_store.similarity_search(query, k=2)
    return " ".join([result.page_content for result in results])
```

```python
# Create a chat prompt template
template = """
You are a customer facing sales and information chatbot for my company Rejolut live on the company website rejolut.com, which focuses on providing tech services. Act as a polite, helpful chatbot that aims to help customers know more about our company Rejolut and try to convince them to try our services. Use the context given for answering. If needed, act like an expert in the areas where we provide service (tech, ai/ml, blockchain, web3)
Respond with 'Sorry, I can't help with that' if you don't know the answer.

We have no confidentiality agreements with our clients, so you can share about the companies we've worked with.

Just answer what the user asks, without giving any other info

Context: {context}

Question: {question}
"""

# Initialize the chat prompt template
chat_prompt = ChatPromptTemplate.from_template(template)
```

```python
# Define the function to generate a response using the context
def query_with_context(user_input):
    context = search_context(user_input)
    prompt_text = chat_prompt.format(context=context, question=user_input)
    messages = [{"role": "user", "content": prompt_text}]
    output = pipe(messages, **generation_args)
    return output[0]['generated_text']
```

```python
# Example usage 1
user_input = "What are Rejolut's core values?"
response = query_with_context(user_input)
print(response)
```
